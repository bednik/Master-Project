```python
import os
import sys
import numpy as np
import nibabel as nib
import matplotlib.pyplot as plt
```

```python
def show_slices(slices):
   """ Function to display row of image slices """
   fig, axes = plt.subplots(1, len(slices))
   for i, slice in enumerate(slices):
       axes[i].imshow(slice.T, cmap="gray", origin="lower")
```

```python
img = nib.load("../VolumeData/MRI/abdomen.nii")
```

```python
arr = img.get_fdata()
arrc = arr.reshape(-1, order='C')
```

```python
slice_0 = arr[26, :, :]
slice_1 = arr[:, 30, :]
slice_2 = arr[:, :, 16]

print("Shape: " + str(arr.shape))
print("Header says c order? " + str(arr.flags['C_CONTIGUOUS']))

show_slices([slice_0, slice_1, slice_2])
plt.suptitle("Fortran order")
```

```python
slice_0 = arrc[26, :, :]
slice_1 = arrc[:, 30, :]
slice_2 = arrc[:, :, 16]

print("Shape: " + str(arrc.shape))
print("Header says c order? " + str(arrc.flags['C_CONTIGUOUS']))

show_slices([slice_0, slice_1, slice_2])
plt.suptitle("C order")
```
